```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
```

```python
COVID_econ_df = pd.read_csv("COVID_econ.csv")
```

```python
indicators = ["SP500", "Dow_Jones_Industrial_Average", "NASDAQ", "Russell2000"]

for indicator in indicators:
    fig, (ax1, ax2, ax3) = plt.subplots(3, sharex=True)
    fig.suptitle(f"{indicator} Performance vs. COVID cases over time")

    ax1.plot(COVID_econ_df["Date"], COVID_econ_df["Total Cases"], color="blue")
    ax1.set_ylabel("cases")

    ax2.plot(COVID_econ_df["Date"], COVID_econ_df[f"Close_{indicator}"], color="orange")
    ax2.set_ylabel(f"Close price {indicator}")

    ax3.plot(COVID_econ_df["Date"], COVID_econ_df[f"Volume_{indicator}"], color="red")
    ax3.set_ylabel(f"Volume {indicator}")

    plt.xlim(min(COVID_econ_df["Date"]), max(COVID_econ_df["Date"]))
    plt.xlabel("Date")
    plt.xticks(rotation="vertical")
    
    correlation = sts.pearsonr(COVID_econ_df["Total Cases"], COVID_econ_df[f"Close_{indicator}"])
    correlation2 = sts.pearsonr(COVID_econ_df["Total Cases"], COVID_econ_df[f"Volume_{indicator}"])
    
    print(f"The correlation between cases and {indicator} closing price is {round(correlation[0],2)}")
    print(f"The correlation between cases and {indicator} volume is {round(correlation2[0],2)}")
```

```python
indicators = ["SP500", "Dow_Jones_Industrial_Average", "NASDAQ", "Russell2000"]


fig, (ax1, ax2, ax3) = plt.subplots(3, sharex=True, figsize=(8,5))
fig.suptitle(f"Economic Indicator Performance vs. COVID cases over time")

ax1.plot(COVID_econ_df["Date"], COVID_econ_df["Total Cases"], color="blue")
ax1.set_ylabel("cases")

for indicator in indicators:
    ax2.plot(COVID_econ_df["Date"], COVID_econ_df[f"Close_{indicator}"], label=f"{indicator}")
    ax2.set_ylabel("Close price")
    ax2.legend(loc="upper left", bbox_to_anchor=(0,2.225))

    ax3.plot(COVID_econ_df["Date"], COVID_econ_df[f"Volume_{indicator}"], label=f"{indicator}")
    ax3.set_ylabel("Volume")

plt.xlim(min(COVID_econ_df["Date"]), max(COVID_econ_df["Date"]))
plt.xlabel("Date")
plt.xticks(rotation="vertical")
```

```python
cse_df = pd.read_csv("construction_unemployment_rate.csv")

ages = ["16yo and over", "16-19yo", "20yo and over", "20-24yo", "25-34yo", "35-44yo", "45-54yo", "55yo and over"]

for age in ages:
    plt.figure()
    plt.plot(cse_df["Date"], cse_df[f"Unemployment Rate, {age}"], label=f"{age}")
    plt.plot(cse_df["Date"], cse_df[f"Unemployment Rate, Men, {age}"], label=f"Men {age}")
    plt.plot(cse_df["Date"], cse_df[f"Unemployment Rate, Women, {age}"], label=f"Women {age}")
    plt.title(f"Unemployment Rate of Men and Women {age}")
    plt.ylabel("Unemployment Rate (%)")
    plt.xlabel("Date")
    plt.legend(loc="best")
    plt.xlim(min(cse_df["Date"]), max(cse_df["Date"]))
    plt.xticks(cse_df["Date"][::12], rotation="vertical")
    plt.tight_layout()

#     plt.savefig(f"output/Unemployment Rate of Men and women {age}.png")
```

```python
ages = ["16yo and over", "16-19yo", "20yo and over", "20-24yo", "25-34yo", "35-44yo", "45-54yo", "55yo and over"]

plt.figure(figsize=(15,6))

for age in ages:
    plt.plot(cse_df["Date"], cse_df[f"Unemployment Rate, {age}"], label=f"{age}")
    plt.title("Unemployment Rate by Age")
    plt.ylabel("Unemployment Rate (%)")
    plt.xlabel("Date")
    plt.legend(loc="best")
    plt.xlim(min(cse_df["Date"]), max(cse_df["Date"]))
    plt.xticks(cse_df["Date"][::12], rotation="vertical")
    plt.tight_layout()
    
    plt.savefig("output/Unemployment Rate by Age.png")
```

```python
race = ["White", "African American", "Asian", "Hispanic or Latino"]

plt.figure(figsize=(8,6))

for x in race:
    plt.plot(cse_df["Date"], cse_df[f"Unemployment Rate, {x}, 16yo and over"], label=f"{x}")
    plt.title(f"Unemployment Rate by race")
    plt.ylabel("Unemployment Rate (%)")
    plt.xlabel("Date")
    plt.legend(loc="best")
    plt.xlim(min(cse_df["Date"]), max(cse_df["Date"]))
    plt.xticks(cse_df["Date"][::12], rotation="vertical")
    plt.tight_layout()
    
plt.savefig("output/Unemployment Rate by race.png")
```

```python
ages = ["16-19yo", "20yo and over"]

race = ["White", "African American", "Hispanic or Latino"]

for age in ages:
    plt.figure(figsize=(15,5))
    plt.plot(cse_df["Date"], cse_df[f"Unemployment Rate, Men, {age}"], label="Men")
    plt.plot(cse_df["Date"], cse_df[f"Unemployment Rate, Women, {age}"], label="Women")
    for x in race:
        plt.plot(cse_df["Date"], cse_df[f"Unemployment Rate, {x}, Men, {age}"], label=f"{x} Men")
        plt.plot(cse_df["Date"], cse_df[f"Unemployment Rate, {x}, Women, {age}"], label=f"{x} Women")
        plt.title(f"Unemployment Rate of {age} by race")
        plt.ylabel("Unemployment Rate (%)")
        plt.xlabel("Date")
        plt.legend(loc="best")
        plt.xlim(min(cse_df["Date"]), max(cse_df["Date"]))
        plt.xticks(cse_df["Date"][::12], rotation="vertical")
        plt.tight_layout()
        
#     plt.savefig(f"output/Unemployment Rate of {age} by race.png")
```
