```python
states_df = pd.read_csv("state_unempoyment_yearly.csv")

fig, ax1 = plt.subplots(sharey=True, figsize=(10,5))

fig.suptitle("State Unemployment Rate by years")

ax1.boxplot([states_df[f"{year}"] for year in states_df.iloc[:,2:-2].columns])
ax1.set_ylabel("Unemployment Rate (%)")
ax1.set_xticklabels(states_df.iloc[:,2:-2].columns)
ax1.set_xlabel("Year")

plt.savefig("output/State Unemployment Rate by Years.png")

plt.show()
```

```python
for year in states_df.iloc[:,2:-2].columns:
    quartiles = states_df[year].quantile([0.25,0.5,0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    upperbound = upperq + 1.5*iqr
    lowerbound = lowerq - 1.5*iqr
    for index, row in states_df.iterrows():
        if row[year] > upperbound:
            print(f"{row['name']} is an upper outlier for {year} | Upperbound: {round(upperbound,2)}, {row['name']}: {round(row[year], 2)}")
        elif row[year] < lowerbound:
            print(f"{row['name']} is a lower outlier for {year} | Lowerbound {round(lowerbound,2)}, {row['name']}: {round(row[year],2)}")
```

```python
states2_df = pd.read_csv("state_unemployment_monthly.csv")

states2_df.drop(["Year", "month"], axis=1, inplace=True)

fig, ax1 = plt.subplots(sharey=True, figsize=(15,5))

fig.suptitle("State Unemployment Rate by state")

ax1.boxplot([states2_df[f"{state}"] for state in states2_df.iloc[:,1:].columns])
ax1.set_ylabel("Unemployment Rate (%)")
ax1.set_xticklabels(states2_df.iloc[:,1:].columns, rotation="vertical")
ax1.set_xlabel("State")

plt.savefig("output/State Unemployment Rate by State.png")

plt.show()
```

```python
locations = states_df[["latitude", "longitude"]]

gmaps.configure(api_key=g_api_key)
fig = gmaps.figure()

max_i = max([max(states_df[year]) for year in states_df.iloc[:,1:-2].columns])

heat_layer = gmaps.heatmap_layer(locations, weights=states_df["2010"],
                                dissipating=False, max_intensity= max_i,
                                point_radius=5)
fig.add_layer(heat_layer)

symbol_layer = gmaps.symbol_layer(locations, fill_color="blue",
                                 stroke_color="blue", scale=2,
                                hover_text = [f"{row['name']}'s Average Unemployment Rate of 2008: {round(row['2010'],2)}%" for index, row in states_df.iterrows()],
                                 info_box_content= [f"{row['name']}'s average unemployment rate of 2008: {row['2010']}%" for index, row in states_df.iterrows()])
fig.add_layer(symbol_layer)

fig
```
