```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
```

```python
level_of_education_df = pd.read_csv("Combined_Ouput_File/level_of_education.csv")
```

```python
plt.figure(figsize=(15,10))
educationsyearframe = level_of_education_df['Date']


collegeeducation = plt.plot(educationsyearframe, level_of_education_df["Bachelor's Degree or Higher"], color='purple')
somecollegeeducation = plt.plot(educationsyearframe, level_of_education_df['Some College'], color='green')
highschooleducation = plt.plot(educationsyearframe, level_of_education_df['High School Graduate'], color='red')
nohighschooleducation = plt.plot(educationsyearframe, level_of_education_df['No High School Diploma'], color='blue')
red_glabel = mpatches.Patch(color='red', label='High School Graduate')
blue_glabel = mpatches.Patch(color='blue', label='No High School Diploma')
green_glabel = mpatches.Patch(color='green', label='Some College')
purple_glabel = mpatches.Patch(color='purple', label="Bachelor's Degree or Higher")
plt.legend(handles=[blue_glabel, red_glabel, green_glabel, purple_glabel])
plt.title('Unemployment impact on Different Education Levels')
plt.ylabel('Unemployment Percentage (Monthly %)')

plt.xlabel('Year')
plt.grid()
plt.xlim('2006-1', '2018-12')
plt.xticks(level_of_education_df['Date'][::12], rotation='vertical')
plt.show()
```

```python
recession_layoffs_data = pd.read_csv("Combined_Ouput_File/recession_layoffs_data.csv")
```

```python
onlayoff = recession_layoffs_data['Loss of Job: On Layoff'].sum()
offlayoff = recession_layoffs_data['Loss of Job: Not on Layoff'].sum()

layofflabels = ['On Layoff', 'Not on Layoff'] 
layoffsplit = [onlayoff, offlayoff]
layoffcolors = ['orange', 'lightskyblue']
plt.figure(figsize=(10, 10))
plt.pie(layoffsplit, labels=layofflabels, colors=layoffcolors,
        autopct="%1.1f%%", startangle=180)
plt.title("Job Loss % On Layoff vs. Not on Layoff (2008 - 2011)")
plt.show()
```
